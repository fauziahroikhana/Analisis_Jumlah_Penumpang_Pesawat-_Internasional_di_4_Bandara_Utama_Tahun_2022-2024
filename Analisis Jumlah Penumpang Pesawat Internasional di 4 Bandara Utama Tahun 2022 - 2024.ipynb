```python
import pandas as pd
import numpy as np

base_path = "/content/drive/MyDrive/DW_PROJECT/"

file_list = [
    "Kualanamu-Medan_2022-2024.xlsx",
    "Soekarno Hatta-Jakarta_2022-2024.xlsx",
    "Juanda-Surabaya_2022-2024.xlsx",
    "Ngurah Rai-Bali_2022-2024.xlsx"
]

for file_name in file_list:
    try:
        print(f"\nMemproses file: {file_name}")

        # Load data
        df = pd.read_excel(base_path + file_name)
        print("Kolom awal:", df.columns.tolist())

        # Hapus kolom Unnamed (jika ada)
        kolom_unnamed = [c for c in df.columns if c.startswith("Unnamed")]
        if kolom_unnamed:
            df = df.drop(columns=kolom_unnamed)
            print(f"Dihapus {len(kolom_unnamed)} kolom Unnamed")

        # Kolom numerik
        exclude_cols = ["Tahun"]
        numeric_cols = [c for c in df.columns if c not in exclude_cols]

        # Replace simbol tidak valid
        df[numeric_cols] = df[numeric_cols].replace("-", np.nan)

        # Konversi ke numerik
        for col in numeric_cols:
            df[col] = pd.to_numeric(df[col], errors="coerce")

        # Drop baris tanpa Tahun
        df = df.dropna(subset=["Tahun"])

        print("Preview setelah cleaning:")
        print(df.head())

        # Simpan
        output_name = "clean_" + file_name
        df.to_excel(base_path + output_name, index=False)

        print(f"File bersih disimpan: {output_name}")

    except FileNotFoundError:
        print(f"File tidak ditemukan: {file_name}")
```

# Feature Engineering

```python
import pandas as pd
import numpy as np

base_path = "/content/drive/MyDrive/DW_PROJECT/"

file_list = [
    "clean_Kualanamu-Medan_2022-2024.xlsx",
    "clean_Soekarno Hatta-Jakarta_2022-2024.xlsx",
    "clean_Juanda-Surabaya_2022-2024.xlsx",
    "clean_Ngurah Rai-Bali_2022-2024.xlsx"]

bulan_order = {
    "Januari": 1, "Februari": 2, "Maret": 3, "April": 4,
    "Mei": 5, "Juni": 6, "Juli": 7, "Agustus": 8,
    "September": 9, "Oktober": 10, "November": 11, "Desember": 12}

for file_name in file_list:
    print(f"\nMemproses FE Tahunan: {file_name}")

    df = pd.read_excel(base_path + file_name)
    df[["Bulan", "Tahun_Angka"]] = df["Tahun"].str.split(" ", expand=True)
    df["Tahun_Angka"] = df["Tahun_Angka"].astype(int)
    df["Bulan_Order"] = df["Bulan"].map(bulan_order)

    # Pastikan numerik
    cols = ["Keberangkatan", "Kedatangan", "Transit"]
    for col in cols:
        df[col] = pd.to_numeric(df[col], errors="coerce").fillna(0)

    # Total penumpang per tahun
    df["Total_Penumpang"] = df["Keberangkatan"] + df["Kedatangan"] + df["Transit"]

    # Rasio pergerakan
    df["Rasio_Keberangkatan"] = df["Keberangkatan"] / df["Total_Penumpang"]
    df["Rasio_Kedatangan"]   = df["Kedatangan"] / df["Total_Penumpang"]
    df["Rasio_Transit"]      = df["Transit"] / df["Total_Penumpang"]

    # Growth Year-on-Year
    df = df.sort_values(["Tahun_Angka", "Bulan_Order"])
    df["Growth_Total_YoY"] = df["Total_Penumpang"].pct_change() * 100

    df["Growth_Keberangkatan_YoY"] = df["Keberangkatan"].pct_change() * 100
    df["Growth_Kedatangan_YoY"]   = df["Kedatangan"].pct_change() * 100
    df["Growth_Transit_YoY"]      = df["Transit"].pct_change() * 100

    output_name = "FE_" + file_name
    df.to_excel(base_path + output_name, index=False)

    print("Feature berhasil ditambahkan:")
    print([
        "Total_Penumpang",
        "Rasio_Keberangkatan",
        "Rasio_Kedatangan",
        "Rasio_Transit",
        "Growth_Total_YoY",
        "Growth_Keberangkatan_YoY",
        "Growth_Kedatangan_YoY",
        "Growth_Transit_YoY"
    ])
```

```python
#Mengurutkan bulan
bulan_order = [
    "Januari","Februari","Maret","April","Mei","Juni",
    "Juli","Agustus","September","Oktober","November","Desember"
]
df["Bulan"] = pd.Categorical(df["Bulan"], categories=bulan_order, ordered=True)

#Pivot: Bulan jadi kolom (nilai = Keberangkatan)
df_pivot = df.pivot_table(
    index="Tahun",        # baris = tahun
    columns="Bulan",      # kolom = bulan
    values="Keberangkatan",
    aggfunc="sum"
)
# Hitung matriks korelasi
correlation_matrix = df_pivot.corr()

# Buat heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="RdBu_r", fmt=".2f", linewidths=0.5)
plt.title('Heatmap Korelasi Keberangkatan antar bulan', fontsize=16)
plt.show()
```
